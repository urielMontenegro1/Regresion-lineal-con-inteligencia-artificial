### **regresión lineal 1**




```python
#importamos librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import linear_model
from sklearn.metrics import mean_absolute_error, mean_squared_error, median_absolute_error, r2_score, explained_variance_score


#frame de la regresión
df=pd.read_csv('df_regresion_lineal_1.cvs')

#print(df)

#extraemos los valores
x=np.asanyarray(df[['x']])
y=np.asanyarray(df[['y']])

model=linear_model.LinearRegression()
model.fit(x,y)

yp=model.predict(x)

#imprimimos las regresions
print('Coeficientes(',model.intercept_,',',model.coef_,')')

#metricas de las regresiones
e1=np.abs(y-yp)
#print('Error por cada predicción',e1.transpose())
e2 = mean_absolute_error(y,yp)
print('MAE=',e2)
e3 = mean_squared_error(y,yp)
print('MSE=',e3)
e4 = median_absolute_error(y,yp)
print('MedAE=',e4)
e5 = r2_score(y,yp)
print('R^2=',e5)
e6 = explained_variance_score(y,yp)
print('EVS=',e6)

#ploteamos las graficas
plt.figure()
plt.grid()
plt.title('Regresión lineal')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y,'r*')
plt.plot(x,yp,'b-',lw=2)
plt.legend(['datos','predicción'])
plt.show()
```

### **Poblacion**

```python
df=pd.read_csv('df_regresion_lineal_pop.cvs')


x=np.asanyarray(df[['year']])
y=np.asanyarray(df[['pop']])

model=linear_model.LinearRegression()
model.fit(x,y)
pp=model.predict(x)
yp=model.predict([[2020]])

print('Coeficientes=',yp)


#metricas
e1=np.abs(y-yp)
#print('Error por cada predicción',e1.transpose())
e2 = mean_absolute_error(y,pp)
print('MAE=',e2)
e3 = mean_squared_error(y,pp)
print('MSE=',e3)
e4 = median_absolute_error(y,pp)
print('MedAE=',e4)
e5 = r2_score(y,pp)
print('R^2=',e5)
e6 = explained_variance_score(y,pp)
print('EVS=',e6)

plt.figure()
plt.grid()
plt.title('Regresión lineal de población 2020')
plt.xlabel('año')
plt.ylabel('población')
plt.plot(x,y,'r*')
plt.plot(x,pp,'b-')
plt.legend(['datos','predicción'])
plt.show()
```
